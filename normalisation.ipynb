### Normalising the data

```python
X = stars.values
X_scaled = scale(X)
X_scaled[:2]
```

#### values after scaling

```python
pd.DataFrame(X_scaled, columns=stars.columns).head()
```

#### mean after scaling

```python
np.round(np.mean(X_scaled, axis=0), decimals=10)
```

#### variance after scaling

```python
variance = np.var(X_scaled, axis=0)
variance_ratio = variance / np.sum(variance)
print(sum(variance_ratio))
print(variance_ratio)
print(variance)
```
